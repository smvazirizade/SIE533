```python
train_df = pd.read_csv('train_withErr.csv')
```

```python
#1
train_df[["Parch", "Survived"]].groupby(['Parch'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#2
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
train_df.head()
train_df['IsAlong'] = 1
train_df
```

```python
train_df.loc[train_df['FamilySize'] ==1,"IsAlong"]=0
train_df[["IsAlong", "Survived"]].groupby(['IsAlong'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#3
print(train_df.Embarked.mode())
train_df.loc[train_df.Embarked.isnull(),"Embarked"]
train_df.loc[train_df.Embarked.isnull(),"Embarked"]='S'
train_df.loc[train_df.Embarked.isnull(),"Embarked"]
train_df[["Embarked", "Survived"]].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#4
train_df['EmbarkedMapped']=train_df['Embarked'].map( {'C': 0, 'Q': 1, 'S': 2} ).astype(int) 
train_df
```
